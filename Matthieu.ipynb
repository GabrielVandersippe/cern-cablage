```python
import numpy as np
import matplotlib.pyplot as plt
im = plt.imread('1005_20UPGM23211816_AfterBonding.jpg')
plt.imshow(im)
```

```python
def norme(p):
    return np.sqrt( (p[0]/255)**2 + (p[1]/255)**2 + (p[2]/255)**2 )

def trace(TEST):
    hauteur, largeur = TEST.shape[0], TEST.shape[1]
    L = []
    for i in range(hauteur):
        p = TEST[i,largeur//2,]
        L.append(norme(p))
    return L

def compte(TEST):
    lastcounted = False
    compteur = 0
    L = trace(TEST)
    for x in L:
        if x > 1.7 and not lastcounted:
            compteur+=1
            lastcounted = True
        elif x < 1.7:
            lastcounted = False
    return compteur

sectionimage = np.array([[ im[i,770 + j,].tolist() for j in range(15) ] for i in range(5504) ]) # section sur laquelle on va compter
plt.figure()
plt.grid()
plt.plot(trace(sectionimage))
```

```python
compte(sectionimage)
```

Prenons la médiane sur plusieurs colonnes de pixels pour être plus sûr du résultat

```python
sections = [    np.array([[ im[i,(750 + j) + 5*shift,].tolist() for j in range(15) ] for i in range(5504) ])    for shift in range(5)  ]
L = []
for s in sections:
    L.append(compte(s))
np.median(L)
```
